Se seleccionarán:
-Bp: que indica la presión arterial debido a que "La presión arterial alta hace que el corazón trabaje más duro y, con el tiempo, puede dañar los vasos sanguíneos por todo el cuerpo. Si los vasos sanguíneos de los riñones se dañan, es posible que dejen de eliminar los desechos y el exceso de líquido del cuerpo."https://www.niddk.nih.gov/health-information/informacion-de-la-salud/enfermedades-rinones/presion-arterial-insuficiencia-renal acompañado de esto, es estudio "Clustering-based spatial analysis (CluSA) framework through graph neural network for chronic kidney disease prediction using histopathology images" en el cual trabajan la presión arterial como una de sus varaibles. chrome-extension://efaidnbmnnnibpcajpcglclefindmkaj/https://www.nature.com/articles/s41598-023-39591-8.pdf. "El valor máximo va de 130 a 139 milímetros de mercurio y el valor mínimo está entre 80 y 89 milímetros de mercurio" https://www.mayoclinic.org/es/diseases-conditions/high-blood-pressure/symptoms-causes/syc-20373410

-Sod: indica los niveles de sodio, esta varible es seleccionada debido a que "Una de las funciones más importantes de los riñones es la de filtrar el sodio, sacándolo del cuerpo en la orina. Cuando los riñones se dañan, no pueden filtrar tan bien como cuando están sanos. Esto puede hacer que el sodio permanezca en el cuerpo y haga que la presión arterial suba." https://www.niddk.nih.gov/-/media/Files/Health-Information/Health-Professionals/Kidney-Disease/SodiumTipsForCKD_Spanish.pdf. "Un nivel normal de sodio en la sangre oscila entre 135 y 145 miliequivalentes por litro (mEq/L)."https://www.mayoclinic.org/es/diseases-conditions/hyponatremia/symptoms-causes/syc-20373711#:~:text=Un%20nivel%20normal%20de%20sodio,debajo%20de%20135%20mEq%2FL.

Se escogieron las variables de: Bp (presión), Sod (nivel de sodio) las razones de estos dos fueron explicadas anteriormente. Las otras variables son:

-Bu: Urea en sangre debido a que "De forma habitual, los riñones filtran la urea, pero cuando no funcionan bien, la cantidad de urea filtrada disminuye y aumenta en sangre" https://www.riojasalud.es/servicios/nefrologia/articulos/interpretando-los-analisis-del-rinon#:~:text=De%20forma%20habitual%2C%20los%20ri%C3%B1ones,alteraci%C3%B3n%20del%20nivel%20de%20conciencia.


-Hemo: indica el nivel de hemoglobina y es escogida debido a que: "La anemia es una complicación frecuente de la enfermedad renal y se asocia con una disminución en la calidad de vida de los pacientes, así como con un aumento de la morbimortalidad y de progresión de la enfermedad."https://www.revistanefrologia.com/es-anemia-enfermedad-renal-cronica-protocolo-articulo-S0211699517301881

-Pot: indica el nivel de potasio y se escogió debido a que: "Las personas con enfermedad renal corren riesgo de tener cifras elevadas de potasio porque los riñones son incapaces de depurar de la sangre el exceso de potasio."https://www.kidneyfund.org/es/viviendo-con-enfermedad-renal/los-problemas-de-salud-causados-por-la-enfermedad-renal/las-cifras-altas-de-potasio-hiperpotasemia-causas-prevencion-y-tratamiento#:~:text=Las%20personas%20con%20enfermedad%20renal%20corren%20riesgo%20de%20tener%20cifras,y%20regresa%20al%20torrente%20sangu%C3%ADneo.

-Al: indica el nivel de albumina en orina y se escogió debido a que: "La albuminuria es un signo de enfermedad renal y significa que el paciente tiene exceso de albúmina en la orina. La albúmina es una proteína que se encuentra en la sangre. Un riñón sano no permite que la albúmina pase de la sangre a la orina, mientras que un riñón dañado sí deja pasar algo de albúmina a la orina."https://www.niddk.nih.gov/health-information/informacion-de-la-salud/enfermedades-rinones/informacion-general/diagnostico/albuminuria-albumina-orina#:~:text=La%20albuminuria%20es%20un%20signo,de%20alb%C3%BAmina%20a%20la%20orina.

```python
data_2=data[["Bp","Sod",	"Bu",	"Hemo",	"Pot", "Al", "Class"]]
```

```python
data_2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Gráfica de Pastel:

```python
fig, ax = plt.subplots()
resultadoGeneral=data_2['Bp'].value_counts(normalize=True)
Nombres=data_2['Bp'].unique()
ax.pie(resultadoGeneral,labels = Nombres,autopct = '%1.1f%%', rotatelabels = True, wedgeprops = {"linewidth": 0.5, "edgecolor": "white"})
ax.set_title('Diagrama de Pastel de la Presión Sanguinea')
```

Mediante la gráfica se puede ver que un porcentaje de 82.2% poseen valores menores a 89 milímetros de mercurio lo que indica Hipertensión de etapa 1 la cual es un factor de riesgo para la insufiencia renal aguda.

Gráfica de Histograma para la albumina:

```python
fig, ax = plt.subplots()
ax.hist(data_2['Al'],edgecolor = "white")
ax.set_title('Histograma para los niveles de Albumina')
```

Mediante la gráfica se puede observar que los niveles de la mayoria de registros se encuentran por debajo de 3.2 que son los valores normales, esto indica una desnutrición asi como problemas renales como lo son síndrome nefrótico y en la insuficiencia renal crónica. "https://www.tuasaude.com/es/examen-de-albumina/"

Gráfica de barras para Bu:

```python
fig, ax = plt.subplots()
contar_2=data_2['Bu'].value_counts(normalize=True)
Nombres_2=data_2['Bu'].unique()
ax.bar( x = Nombres_2, height = contar_2)
ax.set_title('Diagrama de Barras para la Urea en Sangre')
```

Mediante la gráfica se puede observar que la mayor concentracion de valores se encuentran entre 0 y 50, un parte un poco menor de valores, pero aun asi significativos, aumentan de los 60mg/dl lo que indica que debee realizarse una consulta medica y puede indicar problemas renales. https://healthlibrary.brighamandwomens.org/Spanish/RelatedItems/167,urea_nitrogen_serum_es#:~:text=El%20nivel%20normal%20de%20BUN,el%20estado%20de%20los%20ri%C3%B1ones.

Gráfico de Caja para Sod:

```python
sns.set_style("whitegrid")
sns.boxplot(y = "Sod", data = data_2).set_title("Niveles de Sodio")
```

En la gráfica se puede observar que los datos se agrupan principalmente entre los valores de 140 y 130 lo que indican valores normales, unos pocos datos se pueden ver que bajan de los 130, estos indicarian hipanotremia que es un indicativo de problemas renales.

Gráfica Strip plot de Pot:

```python
sns.stripplot( y = data_2["Pot"],
                edgecolor = "black",color = "darkseagreen")
```

En la gráfica se puede observar una concentracion de los valores al rededor de los 5mmol/L, la parte inferior a estos son los registros que tienen niveles normales y la parte superior a estos poseen hiperepotasemia que es un indicativo de problemas renales. Las proporciones de estos registros estan al rededor de 50-50 (niveles normales - niveles elevados).

Gráfica de densidad para Hemo:

```python
sns.kdeplot(x = data_2['Hemo'], color ="red", fill = True)
```

"El rango normal de hemoglobina varía generalmente de 13,2 a 16,6 gramos (g) de hemoglobina por decilitro (dL) de sangre en los hombres y de 11,6 a 15 gramos por decilitro en las mujeres."https://www.mayoclinic.org/es/diseases-conditions/iron-deficiency-anemia/diagnosis-treatment/drc-20355040.

En la gráfica se puede apreciar que la mayoria de los registros se encuentran dentro de esos rangos, los que estan por fuera indican la presencia de anemia que es uno de los factores de riesgo de las enfermedades renales.
