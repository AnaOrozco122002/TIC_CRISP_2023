```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Se escogieron las variables de: chol (colesterol), fbs (nivel de azucar en ayunas) las razones de estos dos fueron explicadas anteriormente. Las otras variables son:

-trestbps: que indica el valor de la presión sanguinea en reposo, se escogió esta variable debido a que: "La hipertensión supone una mayor resistencia para el corazón, que responde aumentando su masa muscular (hipertrofia ventricular izquierda) para hacer frente a ese sobreesfuerzo. Este incremento de la masa muscular acaba siendo perjudicial porque no viene acompañado de un aumento equivalente del riego sanguíneo y puede producir insuficiencia coronaria y angina de pecho." https://fundaciondelcorazon.com/prevencion/riesgo-cardiovascular/hipertension-tension-alta.html .

-La siguiente variable escogida es thalach: (maximo valor de latidos alcanzados) debido a que: "La taquicardia es un ritmo cardíaco irregular o acelerado, generalmente de más de 100 latidos por minuto, que puede llegar hasta 400. A este ritmo elevado, el corazón no puede bombear sangre con oxígeno a tu cuerpo de manera eficiente." y puede deberse a "Suministro deficiente de sangre al corazón debido a una arteriopatía coronaria (arterioesclerosis), enfermedad valvular del corazón, insuficiencia cardíaca, enfermedad muscular del corazón (miocardiopatía), tumores o infecciones." https://www.medtronic.com/es-es/tu-salud/patologias/pulso-acelerado.html.

-La siguiente variable escogida es: exang (angina de esfuerzo) debido a que: "La angina se debe a la mala circulación a través de los vasos sanguíneos en el corazón denominadas arterias coronarias, El miocardio necesita un suministro continuo de oxígeno. Los síntomas de angina pueden ocurren cuando el suministro de sangre al miocardio se reduce. " https://medlineplus.gov/spanish/ency/article/000198.htm esta enfermedad indica un gran riesgo de problemas cardiacos debido a la falta de oxigeno en el corazón.

-La siguiente variable escogida es la edad debido a que "Los cambios que ocurren con la edad pueden aumentar el riesgo de una persona de desarrollar una enfermedad cardíaca. Una de las causas principales de la enfermedad cardíaca es la acumulación de depósitos de grasa en las paredes de las arterias durante muchos años. El cambio más común en el envejecimiento es el aumento de la rigidez de las arterias mayores, llamado arteriosclerosis o endurecimiento de las arterias. La arterioesclerosis causa presión arterial alta o hipertensión, que se vuelve más común a medida que las personas envejecen."https://www.nia.nih.gov/espanol/salud-corazon-envejecimiento#:~:text=El%20cambio%20m%C3%A1s%20com%C3%BAn%20en,medida%20que%20las%20personas%20envejecen.

```python
data_2=data[["age",	"trestbps",	"chol",	"fbs",	"thalach",	"exang", "num"]]
```

```python
data_2
```

Grafica de Pastel de la Azucar en Ayunas:

```python
fig, ax = plt.subplots()
resultadoGeneral=data_2['fbs'].value_counts(normalize=True)
grupos=["Normal","Superior a 120"]
ax.pie(resultadoGeneral,labels = grupos,autopct = '%1.1f%%')
```

Mediante el gráfico se puede observar que un porcentaje del 85%.1 de los registros poseen un nivel de azúcar normal y un porcentaje del 14.9% sufren de azucar alta lo que puede indicar diabetes la cual es un factor de riesgo para enfermedades cardiacas.

Grafica de Histograma con Chol:

```python
fig, ax = plt.subplots()
ax.hist(data_2['chol'],edgecolor = "white")
```

Como se puede observar en el histograma los valores mas repetidos para el colesterol son superiores a los 200 miligramos por decilitro (mg/dL) lo que indica un alto colesterol y aumenta el riesgo de enfermedades cardíacas.

Gráfica de Barras con exang:

```python
fig, ax = plt.subplots()
contar=data_2['exang'].value_counts(normalize=True)
titulos=["Si","No"]
colores = ["#619cff", "#00ba38"]
ax.bar( x = titulos, height = contar,color = colores)
```

Se puee observar mediante el gráfico de barras que la mayoria de las personas registradas poseen una angina de esfuerzo lo que puede indicar que una gran cantidad de la poblacion posee problemas de estrés emocional o algun otro tipo de predisposicion como lo es la diabetes, presión alta, entre otros, los cuales son factores de riesgo para enfermedades cardiacas.

Gráfico de Caja para trestbps:

```python
sns.set_style("whitegrid")
sns.boxplot(y = "trestbps", data = data_2).set_title("Presión Arterial")
```

Mediante el diagrama de caja se puede observar que la media de los valores se encuentra al rededor de 130 y el primer cuartil de los datos se encuentra en 120 lo que muestra que un gran porcentaje de los registros presentan valores altos de presión alterial lo que puede indicar hipertensión que puede provocar, para aquellos pacientes que ya han tenido un problema cardiovascular, la intensificación el daño; De igual manera si no se ha tenido antes problemas del corazón un factor de riesgo.

Gráfica Strip plot de thalach:

```python
sns.stripplot( y = data_2["thalach"],
                edgecolor = "black",color = "darkseagreen")
```

Mediante la concentración de puntos se puede observar que la mayoria de valores de latidos por minutos se encuentra entre 130 y 180 lo que indica valores elevados y la presencia de la taquicardia y que puede ser provocada por el suministro deficiente de sangre al corazón lo que es un indicativo de enfermedades cardiacas.

Gráfica de densidad para la edad:

```python
sns.kdeplot(x = data_2['age'], color ="red", fill = True)
```

Mediante el gráfico de densidad se puede observar que los valores mas repetidos para la edad van entre los 50 y los 65 años, esto indica que no hay tanta predisposición a las enfermedades cardiacas debido a la edad en las personas registradas en el datset. "Los adultos mayores de 65 años tienen más probabilidad que las personas más jóvenes de padecer enfermedades cardiovasculares" https://www.nia.nih.gov/espanol/salud-corazon-envejecimiento#:~:text=Los%20adultos%20mayores%20de%2065,vasos%20sangu%C3%ADneos%20o%20en%20ambos.
